## Naive Bayes

* Quick-and-dirty baseline for classification problems
* Generative classifier
    * Learn a model of the joint probability
* Parameteric model
    * Assumes distribution follows a Gaussian/Multimodal distribution
* Pros
    * Extremely fast
    * Very few tunable parameters
    * Good for very high-dimensional datasets
* Cons
    * May underfit
    
* https://jakevdp.github.io/PythonDataScienceHandbook/05.05-naive-bayes.html

### Gaussian Naive Bayes

* Learns a single Gaussian distribution (Mean and Std) for each class
* Uses the Mean and Std to compute probability of observing features for label 
* How to find the P(Label) and P(Features) for the Bayes Rule?

![](https://jakevdp.github.io/PythonDataScienceHandbook/figures/05.05-gaussian-NB.png)

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, y);
```

```python
rng = np.random.RandomState(0)
Xnew = [-6, -14] + [14, 18] * rng.rand(2000, 2)
ynew = model.predict(Xnew)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
lim = plt.axis()
plt.scatter(Xnew[:, 0], Xnew[:, 1], c=ynew, s=20, cmap='RdBu', alpha=0.1)
plt.axis(lim);
```

```python
# predict_proba
# Columns give the posterior probabilities of the first and second label
yprob = model.predict_proba(Xnew)
yprob[-8:].round(2)
```

### Multinomial Naive Bayes

* Features are assumed to be generated from a simple multinomial distribution
* Describes the probability of observing counts among a number of categories
* Most appropriate for features that represent counts or count rates
* Text Classification! counting words..

```python
from sklearn.datasets import fetch_20newsgroups

data = fetch_20newsgroups()
data.target_names
```

```python
categories = ['talk.religion.misc', 'soc.religion.christian',
              'sci.space', 'comp.graphics']
train = fetch_20newsgroups(subset='train', categories=categories)
test = fetch_20newsgroups(subset='test', categories=categories)
```

```python
print(train.data[5])
```

```python
# TF-IDF vectorizer
# Attach it to a multinomial naive Bayes classifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

```python
model.fit(train.data, train.target)
labels = model.predict(test.data)
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(test.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=train.target_names, yticklabels=train.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

```python
def predict_category(s, train=train, model=model):
    pred = model.predict([s])
    return train.target_names[pred[0]]
```
